```python
import numpy as np # Standard notation
```

```python
from sklearn.datasets import load_iris
dataset = load_iris()['data']

print(dataset.shape)
print(dataset)
```

```python
onedim = [1,2,4,5,7,8]
print(onedim[3])
print(onedim[3:])
print(onedim[::-1])
```

```python
print(dataset[4,1])
print(onedim[:])
print(dataset[:,0])
```

```python
def stdev(data):
    mean = sum(data)/len(data)
    variance = sum([(x-mean)**2 for x in data])
    return (variance/(len(data)-1))**0.5
print(stdev(onedim))
print(stdev(dataset))
```

```python
from time import time

ratios = []
for samples in range(100):
    start1 = time()
    stdev(dataset)
    end1 = time()-start1
    
    start2 = time()
    dataset.std(axis=0)
    end2 = time()-start2
    
    ratios.append(end1/end2)

print('Python lists were',sum(ratios)/len(ratios),'times slower than the numpy way.')
```

# <span style="color:white">The dataset actually looks like this.</span>
![Scatter plot, separated](https://upload.wikimedia.org/wikipedia/commons/thumb/5/56/Iris_dataset_scatterplot.svg/749px-Iris_dataset_scatterplot.svg.png)

```python
names = ['sepal length','sepal width','petal length','petal width']

try_these = [[2,3]]
for i in try_these:
    plt.rcParams['figure.figsize'] = (5,5)
    size_marker = 10000
    plt.scatter(dataset[:50,i[0]],dataset[:50,i[1]],color='red',s=size_marker)
    plt.scatter(dataset[50:100,i[0]],dataset[50:100,i[1]],color='green',s=size_marker)
    plt.scatter(dataset[100:150,i[0]],dataset[100:150,i[1]],color='blue',s=size_marker)
    plt.xlabel(names[i[0]],size=200)
    plt.ylabel(names[i[1]],size=200)
    plt.savefig('petal_width_v_petal_length.svg',bbox_inches='tight')
    plt.show()
```
